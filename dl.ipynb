```python
import numpy as np
def step(x):
  return 1 if x>=0 else 0
class perceptron:
  def __init__(self,weights,bias):
    self.weights=weights
    self.bias=bias
  def predict(self,inputs):
    total=np.dot(self.weights,inputs)+self.bias
    return step(total)
weights=np.array([1,1])
andbias=-1.5
orbias=-0.5
and_gate=perceptron(weights,andbias)
or_gate=perceptron(weights,orbias)
print("AND GATE")
for i in [(0,0),(0,1),(1,0),(1,1)]:
  print(i,"->",and_gate.predict(np.array(i)))
print("OR GATE")
for i in [(0,0),(0,1),(1,0),(1,1)]:
  print(i,"->",or_gate.predict(np.array(i)))
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class XOR:
    def __init__(self):
        self.w_or = np.array([1, 1])
        self.b_or = -0.5
        self.w_and = np.array([1, 1])
        self.b_and = -1.5
        self.w_out = np.array([1, -2])
        self.b_out = -0.5

    def predict(self, x):
        h1 = step(np.dot(self.w_or, x) + self.b_or)
        h2 = step(np.dot(self.w_and, x) + self.b_and)
        output = step(self.w_out[0]*h1 + self.w_out[1]*h2 + self.b_out)
        return output

xorgate = XOR()
print("XOR GATE")
for i in ([0,0], [0,1], [1,0], [1,1]):
    print(i, "->", xorgate.predict(np.array(i)))
print("XNOR GATE")
for i in ([0,0], [0,1], [1,0], [1,1]):
    print(i, "->", step(np.dot(xorgate.predict(np.array(i)),-1)+0.5))


```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, AveragePooling2D, Input, concatenate, BatchNormalization, Activation, Add
from tensorflow.keras.regularizers import l2

def build_lenet5(input_shape=(32, 32, 1), num_classes=10):
    model = Sequential([
        Conv2D(filters=6, kernel_size=(5, 5), activation='tanh', input_shape=input_shape),
        AveragePooling2D(pool_size=(2, 2)),
        Conv2D(filters=16, kernel_size=(5, 5), activation='tanh'),
        AveragePooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(units=120, activation='tanh'),
        Dense(units=84, activation='tanh'),
        Dense(units=num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("\n--- LeNet-5 Model Summary ---")
    model.summary()
    return model

def build_alexnet(input_shape=(28, 28, 1), num_classes=10):
    # Simplified AlexNet for MNIST
    model = Sequential([
        Conv2D(filters=32, kernel_size=(3, 3), strides=(1, 1), activation='relu', input_shape=input_shape, kernel_regularizer=l2(0.0005)),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
        Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu', kernel_regularizer=l2(0.0005)),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
        Conv2D(filters=128, kernel_size=(3, 3), padding='same', activation='relu', kernel_regularizer=l2(0.0005)),
        Flatten(),
        Dense(units=512, activation='relu', kernel_regularizer=l2(0.0005)),
        Dropout(0.5),
        Dense(units=num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("\n--- Simplified AlexNet Model Summary (for MNIST) ---")
    model.summary()
    return model


def build_zfnet(input_shape=(28, 28, 1), num_classes=10):
    # Simplified ZFNet for MNIST
    model = Sequential([
        Conv2D(filters=32, kernel_size=(3, 3), strides=(1, 1), activation='relu', input_shape=input_shape),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
        Conv2D(filters=64, kernel_size=(3, 3), padding='same', activation='relu'),
        MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
        Conv2D(filters=128, kernel_size=(3, 3), padding='same', activation='relu'),
        Flatten(),
        Dense(units=256, activation='relu'),
        Dropout(0.5),
        Dense(units=num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("\n--- Simplified ZFNet Model Summary (for MNIST) ---")
    model.summary()
    return model

def build_vgg16(input_shape=(28, 28, 1), num_classes=10):
    # Simplified VGG-16 for MNIST
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        Conv2D(32, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2), strides=(2, 2)),

        Conv2D(64, (3, 3), activation='relu', padding='same'),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2), strides=(2, 2)),

        Conv2D(128, (3, 3), activation='relu', padding='same'),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2), strides=(2, 2)),

        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("\n--- Simplified VGG-16 Model Summary (for MNIST) ---")
    model.summary()
    return model

def inception_block(x, filters_1x1, filters_3x3_reduce, filters_3x3, filters_5x5_reduce, filters_5x5, filters_pool_proj):
    conv_1x1 = Conv2D(filters_1x1, (1, 1), padding='same', activation='relu')(x)

    conv_3x3 = Conv2D(filters_3x3_reduce, (1, 1), padding='same', activation='relu')(x)
    conv_3x3 = Conv2D(filters_3x3, (3, 3), padding='same', activation='relu')(conv_3x3)

    conv_5x5 = Conv2D(filters_5x5_reduce, (1, 1), padding='same', activation='relu')(x)
    conv_5x5 = Conv2D(filters_5x5, (5, 5), padding='same', activation='relu')(conv_5x5)

    pool_proj = MaxPooling2D((3, 3), strides=(1, 1), padding='same')(x)
    pool_proj = Conv2D(filters_pool_proj, (1, 1), padding='same', activation='relu')(pool_proj)

    output = concatenate([conv_1x1, conv_3x3, conv_5x5, pool_proj], axis=-1)
    return output

def build_googlenet(input_shape=(28, 28, 1), num_classes=10):
    input_tensor = Input(shape=input_shape)

    x = Conv2D(32, (3, 3), strides=(1, 1), padding='same', activation='relu')(input_tensor)
    x = MaxPooling2D((2, 2), strides=(2, 2))(x)

    x = inception_block(x, 16, 16, 32, 4, 8, 8)
    x = MaxPooling2D((2, 2), strides=(2, 2))(x)

    x = inception_block(x, 32, 32, 64, 8, 16, 16)
    x = MaxPooling2D((2, 2), strides=(2, 2))(x)

    x = AveragePooling2D((2, 2))(x)
    x = Flatten()(x)
    x = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=input_tensor, outputs=x)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("\n--- Simplified GoogleNet (for MNIST) Model Summary ---")
    model.summary()
    return model

def residual_block(x, filters, kernel_size=(3, 3), stride=1, activate_before_res=False):
    shortcut = x
    if activate_before_res:
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        shortcut = Conv2D(filters, (1, 1), strides=stride, padding='same')(shortcut)

    x = Conv2D(filters, kernel_size, strides=stride, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(filters, kernel_size, padding='same')(x)
    x = BatchNormalization()(x)

    x = Add()([shortcut, x])
    x = Activation('relu')(x)
    return x

def build_resnet(input_shape=(28, 28, 1), num_classes=10):
    input_tensor = Input(shape=input_shape)

    x = Conv2D(32, (3, 3), strides=(1, 1), padding='same', activation='relu')(input_tensor)
    x = BatchNormalization()(x)
    x = MaxPooling2D((2, 2), strides=(2, 2))(x)

    x = residual_block(x, 32)
    x = residual_block(x, 32)
    x = MaxPooling2D((2, 2), strides=(2, 2))(x)

    x = residual_block(x, 64, stride=1, activate_before_res=True)
    x = residual_block(x, 64)

    x = AveragePooling2D(pool_size=(2, 2))(x)
    x = Flatten()(x)
    x = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=input_tensor, outputs=x)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("\n--- Simplified ResNet (for MNIST) Model Summary ---")
    model.summary()
    return model

lenet_model = build_lenet5()
alexnet_model = build_alexnet()
zfnet_model = build_zfnet()
vgg16_model = build_vgg16()
googlenet_model = build_googlenet()
resnet_model = build_resnet()
```

## Training VGG-16 on MNIST

**Note:** VGG-16 is a very deep network typically used for images like ImageNet (224x224x3). It is highly unlikely to run successfully or provide meaningful results on 28x28x1 MNIST images without substantial architectural modifications to accommodate the small input size, as many pooling layers would reduce the feature maps to zero dimensions very quickly.

```python
import matplotlib.pyplot as plt

print("\nDemonstrating CNN training on MNIST using Simplified VGG-16...")

# Adjust input_shape and num_classes for MNIST
vgg16_model = build_vgg16(input_shape=(28, 28, 1), num_classes=num_classes)

# Train the model
history = vgg16_model.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test), verbose=1)
loss, accuracy = vgg16_model.evaluate(x_test, y_test, verbose=0)
print(f"\nSimplified VGG-16 Test Accuracy: {accuracy:.4f}")

# Plotting training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Simplified VGG-16 Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Simplified VGG-16 Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
